```python
import numpy as np
import glob
import os
```

```python
model_path = 'BigGAN_I128_hdf5_seed0_Gch96_Dch96_bs128_nDa8_nGa8_Glr1.0e-04_Dlr4.0e-04_Gnlinplace_relu_Dnlinplace_relu_Ginitortho_Dinitortho_Gattn64_Dattn64_Gshared_hier_ema'
```

```python
g_log = '../logs/%s/G_loss.log'%model_path
fo = open(g_log, "r")
line = fo.readlines()
iters = [int(l.split(':')[0]) for l in line]
g_loss = [float(l.split(':')[1].strip()) for l in line]
print(g_loss[0], iters[0])
fo.close()
w = np.ones((20,))
MA = np.convolve(w, g_loss, mode='same')
print(MA.shape, len(iters))
plot(iters[10:-10],MA[10:-10])
print(np.max(iters))
```

```python
D_log_fake = '../logs/%s/D_loss_fake.log'%model_path
fo = open(D_log_fake, "r")
line = fo.readlines()
iters = [int(l.split(':')[0]) for l in line]
d_loss = [float(l.split(':')[1].strip()) for l in line]
fo.close()
w = np.ones((20,))
MA = np.convolve(w, d_loss, mode='same')
print(MA.shape, len(iters))
plot(iters[10:-10],MA[10:-10])

# plot(iters[5:],d_loss[5:])
```

```python
def find_classes(dir):
    classes = [d for d in os.listdir(dir) if os.path.isdir(os.path.join(dir, d))]
    classes.sort()
    class_to_idx = {classes[i]: i for i in range(len(classes))}
    return classes, class_to_idx
data_dir = '../data/ADE15c/'
cls, cls2id = find_classes(data_dir)
```
